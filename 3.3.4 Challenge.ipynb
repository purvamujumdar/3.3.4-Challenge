```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/purva/Desktop/Data Science Boot Camp/table_crime data_csv.csv', header = 1)
```

```python
df.head()
```

```python

df.drop(['Rape\n(revised\ndefinition)1','Arson3', 'Unnamed: 13'], axis=1)
```

```python
df['Property\ncrime']=df['Property\ncrime'].str.replace(',','')
df['Property\ncrime']=df['Property\ncrime'].astype(int)
df['Population'] = (df['Population']).str.replace(',','')
df['Population'] = (df['Population']).astype(int)
df['Robbery'] = (df['Robbery']).str.replace(',','')
df['Robbery'] = (df['Robbery']).astype(int)
df['Larceny-\ntheft'] = (df['Larceny-\ntheft']).str.replace(',','')
df['Larceny-\ntheft'] = (df['Larceny-\ntheft']).astype(int)
df['Burglary'] = (df['Burglary']).str.replace(',','')
df['Burglary'] = (df['Burglary']).astype(int)
```

```python
df['feature_Murder'] = df['Murder and\nnonnegligent\nmanslaughter']
df.loc[df['feature_Murder'] > 0, 'feature_Murder'] = 1
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model
lr = LogisticRegression(C=1e9)
```

```python
X= df[['Population','Robbery', 'Property\ncrime', 'Burglary']]
Y = df['feature_Murder']
```

```python
model1 = lr.fit(X,Y)
```

```python
print('Coefficients')
print(model1.coef_)
print(model1.intercept_)
pred_y_sklearn = lr.predict(X)

print('\n Accuracy')
print(pd.crosstab(pred_y_sklearn, Y))

print('\n Percentage accuracy')
print(lr.score(X, Y))
```

## Ridge Regression

```python
rr = linear_model.Ridge(alpha = 10, fit_intercept = False)
model2 = rr.fit(X,Y)
print('Coefficients')
print(model2.coef_)
print(model2.intercept_)
pred_y_sklearn = rr.predict(X)

##print('\n Accuracy')
##print(pd.crosstab(pred_y_sklearn, Y))

print('\n Percentage accuracy')
print(rr.score(X, Y))
```

## Lasso Regression

```python
lass = linear_model.Lasso(alpha = 0.40)
model3 = lass.fit(X,Y)
print('Coefficients')
print(model3.coef_)
print(model3.intercept_)
pred_y_sklearn = lass.predict(X)

##print('\n Accuracy')
##print(pd.crosstab(pred_y_sklearn, Y))

print('\n Percentage accuracy')
print(lass.score(X, Y))
```
